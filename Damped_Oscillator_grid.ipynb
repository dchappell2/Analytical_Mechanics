### Plotting the solution to the underdamped oscillator

Solution:
$$
x(t)=e^{-\beta t} [A\cos(\omega_1 t) + B\sin(\omega_1 t)]
$$
where
* $A=x_0$
* $B=(v_0+\beta x_0)/\omega_1$
* $x_0 =$ initial position
* $v_0=$ initial velocity
* $\beta = $ damping constant
* $\omega_0 = $ natural frequency

This example plots 6 solutions with different damping parameters. The plots appear as a grid of subplots.

```python
import matplotlib.pyplot as plt
import numpy as np

########  parameters
tmax   = 50     # maximum time
N      = 500    # number of data points
omega0 = 3      # natural frequency of oscillations

beta_array = [.01,.02,.05,.1,.2,.5]     # array of damping coefficients

########  initial conditions
x0 = 1          # initial position
v0 = 0          # initial velocity

# define the size of the plot grid
nPlots_x = 2    # two plots horizontal
nPlots_y = 3    # three plots horizontal

# create a matrix of plots with nPlots rows and 1 column
fig, ax = plt.subplots(nrows=nPlots_y,ncols=nPlots_x,figsize=(8,10))
ax_flat = ax.flatten()

########  function to plot underdamped solution
def plot_solution(title,ax):
  ax.plot(t, x)                          # plot solution
  ax.set_xlabel('t')                         # label the x and y axes
  ax.set_ylabel('x')
  ax.set_title(title)   # give the plot a title

k = 0
for beta in beta_array:

  ########  underdamped oscillations
  omega1 = np.sqrt(omega0**2 - beta**2)
  A      = x0
  B      = (v0 + beta*x0)/omega1

  ########  underdamped solution
  t = np.linspace(0,tmax,N)
  x = np.exp(-beta*t) * (A*np.cos(omega1*t) + B*np.sin(omega1*t))

  # call the plot function
  title = rf"$\beta=${beta}, $\omega_0=${omega0}"
  plot_solution(title,ax_flat[k])
  k += 1

plt.show()                              # display the plot
```
